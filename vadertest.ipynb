```python
import nltk
import pandas as pd
```

```python
nltk.download('vader_lexicon') 
from nltk.sentiment.vader import SentimentIntensityAnalyzer
sid = SentimentIntensityAnalyzer()
df = pd.read_csv('C:/Users/hares/singaporesub/Data/initial_comments_extract.csv')
```

```python
df
```

```python
df['sentiments'] = df['body'].apply(lambda body: sid.polarity_scores(body))

df.head()
```

```python
df.sentiments.head()
```

```python
df['compound']  = df['sentiments'].apply(lambda score_dict: score_dict['compound'])

df.head()
```

```python
df['comp_score'] = df['compound'].apply(lambda c:  'neu' if c>=0 and c<=0.35 else 'pos' if c> 0.35 else 'neg')

df.head(20)
```
